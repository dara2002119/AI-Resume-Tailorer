```python
prompt_template = lambda resume_string, jd_string : f"""
You are a professional resume optimization expert specializing in tailoring resumes to specific job descriptions. Your goal is to optimize my resume and provide actionable suggestions for improvement to align with the target role.

### Hard Requirements (One Page Rule)
- The **Tailored Resume** MUST fit on **one page** when rendered in Word using **Times New Roman 11 pt** with ~0.5–0.75" margins.
- **No sections may be removed**. All core sections must remain:
  - Contact information
  - Education
  - Technical Skills
  - Work Experience
  - Projects (if originally present)
- Sections may be **condensed, merged, or shortened** but never dropped.
- Use these limits to guarantee one page:
  - **Max 540 words** in the Tailored Resume section.
  - **Max 12 bullets total** across all roles/projects (2–3 per role/project).
  - **Max 3 roles** in Work Experience.
  - Compact formatting (comma-separated skills, concise dates like "Jan–May 2025", city/state only, no street address).
- If your draft exceeds ANY limit, you MUST **redo and resubmit a shorter version** until it satisfies the limits.

### Guidelines
1. **Relevance**
   - Prioritize experiences, skills, and achievements most relevant to the job description.
   - Condense but retain all sections.
   - Limit work experience to the 2–3 most relevant roles.
   - Limit bullet points under each role/project to 2–3 of the strongest, most impactful results.

2. **Action-Driven Results**
   - Use **strong action verbs**.
   - Include **quantifiable results** (metrics, % improvements, time/cost savings).

3. **Keyword Optimization**
   - Naturally integrate **keywords and phrases from the job description** for ATS alignment.

4. **Formatting**
   - Output the **Tailored Resume** in **clean Markdown** (no tables).
   - Use compact formatting: comma-separated skills, tight bullets, concise wording.
   - Avoid redundant phrases or padding.

5. **Additional Suggestions (If Gaps Exist)**
   - If the resume doesn’t fully align, suggest:
     1. Additional technical or soft skills to add.
     2. Relevant certifications or courses to pursue.
     3. Projects or experiences to highlight or develop.
   - The one-page limit applies **only to the Tailored Resume**. The **Additional Suggestions** section may exceed one page.

### Self-Check & Retry (MANDATORY)
- After drafting the Tailored Resume, **estimate**:
  - Word count (<= 540), total bullets (<= 12), roles shown (<= 3).
- Verify that **all required sections are present**.
- If any limit is exceeded or a section is missing, **revise and resubmit a shorter version**.
- Include a hidden note as an HTML comment at the very end with your counts, e.g.:
  <!-- length-check: words=512 bullets=11 roles=3 sections=all OK -->

---

### Input
- **My resume**:
{resume_string}

- **The job description**:
{jd_string}

---

### Output
1. **Tailored Resume**  
   - A **Markdown-formatted resume** aligned with the job description.  
   - Emphasizes relevant experience and measurable results.  
   - Integrates ATS-friendly keywords.  
   - **One-page maximum (see limits above, and all sections retained).**

2. **Additional Suggestions** (if applicable)  
   - Recommended skills to add.  
   - Certifications or courses to strengthen candidacy.  
   - Projects or experiences to improve alignment.

(Remember: If the Tailored Resume exceeds one page, omits a required section, or breaks limits, redo and resubmit a shorter version.)
"""
```

```python
prompt = prompt_template(resume_string, jd_string)
```

```python
client = OpenAI(api_key=my_sk)

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Expert resume writer"},
        {"role": "user", "content": prompt}
         ],
    temperature = 0.7
)
response_string = response.choices[0].message.content
```

```python
response_list = response_string.split("## Additional Suggestions")
```

```python
display(Markdown(response_list[0]))
```

```python
output_file = "resumes/resume_new.md"

with open(output_file, "w", encoding="utf-8") as file:
    file.write(response_list[0])
```

```python
import os, re
import pypandoc
from docx import Document
from docx.shared import Inches, Pt, RGBColor
from docx.oxml import OxmlElement
from docx.oxml.ns import qn
from docx.enum.text import WD_ALIGN_PARAGRAPH
from IPython.display import Markdown, display

# ===== 1) Create a compact reference .docx template =====
ref_path = "resumes/resumes_output/resume_reference.docx"
os.makedirs(os.path.dirname(ref_path), exist_ok=True)

doc = Document()

# Page margins
for s in doc.sections:
    s.top_margin = Inches(0.5)
    s.bottom_margin = Inches(0.5)
    s.left_margin = Inches(0.75)
    s.right_margin = Inches(0.75)

def set_style_font(style_obj, font_name="Times New Roman"):
    style_obj.font.name = font_name
    try:
        rPr = getattr(style_obj._element, "rPr", None)
        if rPr is None:
            rPr = OxmlElement('w:rPr')
            style_obj._element.append(rPr)
        rFonts = getattr(rPr, "rFonts", None)
        if rFonts is None:
            rFonts = OxmlElement('w:rFonts')
            rPr.append(rFonts)
        rFonts.set(qn('w:eastAsia'), font_name)
    except Exception:
        pass

# Normal style
normal = doc.styles["Normal"]
set_style_font(normal, "Times New Roman")
normal.font.size = Pt(11)
pf = normal.paragraph_format
pf.line_spacing = 1.0
pf.space_before = Pt(0)
pf.space_after = Pt(0.5)

# Headings
def tune_heading(name, size_pt, color_rgb=None):
    st = doc.styles[name]
    set_style_font(st, "Times New Roman")
    st.font.size = Pt(size_pt)
    if color_rgb is not None:
        st.font.color.rgb = RGBColor(*color_rgb)
    sp = st.paragraph_format
    sp.space_before = Pt(0)
    sp.space_after = Pt(0.25)
    sp.line_spacing = 1.0
    sp.keep_with_next = True

# Built-in heading styles (not centered)
tune_heading("Heading 1", 14)
tune_heading("Heading 2", 12)
tune_heading("Heading 3", 11)

# Bullets = "List Paragraph"
list_para = doc.styles["List Paragraph"]
set_style_font(list_para, "Times New Roman")
list_para.font.size = Pt(11)
lp = list_para.paragraph_format
lp.space_before = Pt(0)
lp.space_after = Pt(0.25)
lp.line_spacing = 1.0
lp.left_indent = Inches(0.18)
lp.first_line_indent = Inches(0)

# Custom NameHeading (blue, centered)
name_heading = doc.styles.add_style("NameHeading", 1)
name_heading.base_style = normal
set_style_font(name_heading, "Times New Roman")
name_heading.font.size = Pt(14)
name_heading.font.color.rgb = RGBColor(0, 102, 204)
nh = name_heading.paragraph_format
nh.alignment = WD_ALIGN_PARAGRAPH.CENTER
nh.space_before = Pt(0)
nh.space_after = Pt(0.25)
nh.line_spacing = 1.0
nh.keep_with_next = True

# Custom Contact style (centered, 11pt)
contact = doc.styles.add_style("Contact", 1)
contact.base_style = normal
set_style_font(contact, "Times New Roman")
contact.font.size = Pt(11)
cp = contact.paragraph_format
cp.alignment = WD_ALIGN_PARAGRAPH.CENTER
cp.space_before = Pt(0)
cp.space_after = Pt(0.25)
cp.line_spacing = 1.0
cp.keep_with_next = True

# Widow control
def set_para_property(style_obj, prop_tag):
    pPr = style_obj._element.get_or_add_pPr()
    el = OxmlElement(prop_tag)
    pPr.append(el)

for sty in (normal, doc.styles["Heading 1"], doc.styles["Heading 2"], 
            doc.styles["Heading 3"], list_para, name_heading, contact):
    set_para_property(sty, 'w:widowControl')

doc.save(ref_path)

# ===== 2) Preprocess MD =====
md_text = response_list[0]

# Strip accidental code-fence like "'''markdown"
md_text = re.sub(r"(?i)^\s*([`']{3,})\s*markdown\s+", "", md_text, count=1)
lines = md_text.splitlines()
if lines and re.match(r"(?i)^\s*([`']{3,})\s*markdown\s*$", lines[0]):
    lines = lines[1:]
    while lines and re.match(r"^\s*([`']{3,})\s*$", lines[-1]):
        lines.pop()
    md_text = "\n".join(lines)
md_text = md_text.lstrip()

# Replace tabs with spaces
md_text = md_text.replace("\t", " ")

md_lines = md_text.splitlines()
h_idx = None
for i, line in enumerate(md_lines):
    if re.match(r'^\s*#{1,6}\s+\S', line):  # first heading
        h_idx = i
        break

if h_idx is not None:
    # Center the first heading
    heading_text = re.sub(r'^\s*#{1,6}\s+', '', md_lines[h_idx]).strip()
    md_lines[h_idx] = f'::: {{custom-style="NameHeading"}}\n{heading_text}\n:::'
    # Also center the very next non-empty line (contact info)
    j = h_idx + 1
    while j < len(md_lines) and md_lines[j].strip() == "":
        j += 1
    if j < len(md_lines):
        contact_line = md_lines[j].strip()
        md_lines[j] = f'::: {{custom-style="Contact"}}\n{contact_line}\n:::'
    md_text = "\n".join(md_lines)

# Remove trailing spaces in bullets
md_text = re.sub(r'(?m)^(\s*[-*]\s.*?)[ \t]{2,}$', r'\1', md_text)
# Collapse extra blank lines
md_text = re.sub(r'\n{3,}', '\n\n', md_text)

# ===== 3) Convert to DOCX =====
output_docx_file = "resumes/resumes_output/resume_one_page.docx"
os.makedirs(os.path.dirname(output_docx_file), exist_ok=True)

pypandoc.convert_text(
    md_text,
    to="docx",
    format="md",
    outputfile=output_docx_file,
    extra_args=["--standalone", f"--reference-doc={ref_path}"]
)

print(f"✅ Exported 1-page-optimized Word doc: {output_docx_file}")

# Optional suggestions
if len(response_list) > 1:
    display(Markdown("## Additional Suggestions\n" + response_list[1]))
```
